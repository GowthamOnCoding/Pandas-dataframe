# Dataframe-Filtering using between()

```python
import pandas as pd
emp_df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
emp_df["Gender"]=emp_df["Gender"].astype("category")
emp_df["Senior Management"]=emp_df["Senior Management"].astype("category")
emp_df["Team"]=emp_df["Team"].astype("category")
emp_df.head()
```

```python
#Lets filter the datframe where bonus percent is between 17 to 19 %
#Remember both the bounds in between method is inclusive
bon_mask=emp_df["Bonus %"].between(17,19)
emp_df[bon_mask].head()
```

```python
#Lets filter where Last Login Time is between 06:00AM and 11:00 AM
time_mask=emp_df["Last Login Time"].between("06:00AM","11:00AM")
emp_df[time_mask].head()
```

```python
#Filter for salary between 60000 and 90000
sal_mask=emp_df["Salary"].between(60000,90000)
emp_df[sal_mask].head(20)
```
