# Dataframe filtering based on single condition

```python
import pandas as pd
emp_df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
#Converting gender,Senior Management,Team to category type
emp_df["Gender"]=emp_df["Gender"].astype("category")
emp_df["Senior Management"]=emp_df["Senior Management"].astype("category")
emp_df["Team"]=emp_df["Team"].astype("category")
emp_df.head(5)
```

```python
emp_df["Team"]=="Marketing"
```

```python
#Now we have to pass this boolean series to data frame filtering
emp_df[emp_df["Team"]=="Marketing"].head()
#We can see that now it is returning the rows which belongs to Team Marketing
```

```python
#We can do the above operation in a elegant way
#Assign a boolean series to a variable
mask=emp_df["Team"]=="Marketing"
#Pass this varable for filtering
emp_df[mask].head()
```

```python
#Filter by salary where salary is greater than 100000
mask=emp_df["Salary"]>100000
emp_df[mask].head()
```

```python
#Filter the datframe where start date is greater than jan 25 1993
mask=emp_df["Start Date"]>"1993-01-25"
emp_df[mask].head()
```
