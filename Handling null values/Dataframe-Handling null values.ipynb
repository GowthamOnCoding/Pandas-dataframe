# Handling null values

```python
import pandas as pd
nba=pd.read_csv("nba.csv")
#Previewing the top 5 rows to check for null values
#We can see that third row last column is having null value
nba.head(5)
```

```python
nba.tail(5)
#We can see that last row in data set is completely null
```

```python

#We can see that after applying dropna method, complete third row is removed since one column value is having null value
nba.dropna().head(5)
```

```python
nba.dropna().tail(5)
#Last row is also removed
```

```python
#To remove a row only if all the column values are null, we can modify how parameter in dropna method
na_remove=nba.dropna(how="all")
na_remove.head()
#We can see that third row is not removed since it has null value in only one column
```

```python
na_remove.tail()
#Last row in dataset is completely removed since it had null values in all columns
```

```python
#To remove a column having null value
nba_upd=nba.head(5)
nba_upd
```

```python
nba_upd.dropna(axis=1).head()
#Above procedure removes a complete column even if it have only one null value, so salary column removed
```

```python
nba_upd.dropna(axis=1,how="all")
#Above method call will remove the column if that column is full of null values
```

```python
#To remove rows by checking null values in a particular column
nba.dropna(subset=["Salary"]).head()
#Third row is removed however 4th and 5th row is retained though it is having null values in the college column
```

```python
# Handling null values using fillna()
#Applying fillna to the whole dataset
nba.fillna(0).head()
#In this context, datframe columns are not consistent rather they are different types, so applying fillna method
#to the whole dataframe does not work here, we can see null value in the college column also replaced by 0 which does not make any sense
```

```python
nba["Salary"].fillna(0,inplace=True)
nba.head()
#We can see now that only salary column null values are replaced by 0
#It is better to extract a series apply fillna to the series if you are satisfied, 
#use inplace parameter to make permanent changes to the dataframe
```

```python
nba["College"].fillna("No college",inplace=True)
nba.head()
#Now null values in college column is replaced by the value No college
```
