# Create random sample

```python
import pandas as pd
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
#To get random rows sample
bond.sample()
```

```python
#TO get 5 rows as sample
bond.sample(n=5)
```

```python
#We can mention the fraction using frac parameter
bond.sample(frac=0.10)
```

```python
#To get column sample
bond.sample(axis=1)
```

```python
bond.sample(n=2,axis=1)
```

```python
bond.sample(axis=1,frac=0.1)
```
