# Retrieving rows using iloc

```python
import pandas as pd
bond=pd.read_csv("jamesbond.csv")
bond.head()
```

```python
bond.loc[5]
```

```python
bond.iloc[5]
```

```python
#Change the index column
bond.set_index("Film",inplace=True)
bond.head()
```

```python
#Retrieving multiple rows
bond.iloc[[5,10]]
```

```python
#Retrieve a range
bond.iloc[:10]
```
