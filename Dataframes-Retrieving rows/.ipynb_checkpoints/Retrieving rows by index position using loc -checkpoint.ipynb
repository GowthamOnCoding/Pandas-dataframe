# Retrieving rows by index position using loc

```python
import pandas as pd
bond=pd.read_csv("jamesbond.csv")
#Make Film as index column
bond.set_index("Film",inplace=True)
bond.sort_index(inplace=True)
bond.head()
```

```python
#Retreiving by index lable
bond.loc["Casino Royale"]
```

```python
#Retrieving range of values
bond.loc["Die Another Day":"Thunderball"]
#Please note that thundrbell also inclusive in this operation whereas in most of the python list operation it is exclusive
```

```python
#Trying to retreive a index label which is not present
bond.loc["Gold Bond"]
#We will get error if we try to retreive a row which is not present
```

```python
#Trying to fetch multiple index labels
bond.loc[["Spectre","Skyfall","Gold Bond"]]
```
