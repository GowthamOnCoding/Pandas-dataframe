# Dataframe multiple condition filtering

```python
import pandas as pd
emp_df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
emp_df["Gender"]=emp_df["Gender"].astype("category")
emp_df["Senior Management"]=emp_df["Senior Management"].astype("category")
emp_df["Team"]=emp_df["Team"].astype("category")
emp_df.head(3)
```

```python
#Boolean series for gender
gen_mask=emp_df["Gender"]=="Female"
team_mask=emp_df["Team"]=="Finance"
#Filtering the dataframe using &
emp_df[gen_mask & team_mask].head()
```

```python
#Filtering using or (|)
#Lets filter the dataframe where salary is greater than 150000 or Bonus % is >19.5
sal_mask=emp_df["Salary"]>110000
bon_mask=emp_df["Bonus %"]>19.5
emp_df[sal_mask|bon_mask].head()
```

```python
#Lets combine both | and &
#Lets try to filter where First Name is Thomas and Salary is greater than 100000 or Senior Management is true
name_mask=emp_df["First Name"]=="Thomas"
sal_mask=emp_df["Salary"]>100000
sm_mask=emp_df["Senior Management"]
#Don't forget to mention the parenthesis otherwise python will get confuse about the evaluation order
emp_df[(name_mask&sal_mask)|sm_mask].tail(15)
```
