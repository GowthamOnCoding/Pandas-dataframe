# Filtering using where() method

```python
import pandas as pd
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.head()
```

```python
#Generate boolean series for budget >6
mask=bond["Budget"]>50

bond.where(mask)
```

```python
#We can have multiple condition inside where()
mask2=bond["Director"]=="Sam Mendes"
mask2
bond.where(mask & mask2)
```
