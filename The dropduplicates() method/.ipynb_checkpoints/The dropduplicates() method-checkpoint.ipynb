# The dropduplicates() method

```python
import pandas as pd
emp_df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
emp_df["Gender"]=emp_df["Gender"].astype("category")
emp_df["Senior Management"]=emp_df["Senior Management"].astype("category")
emp_df["Team"]=emp_df["Team"].astype("category")
emp_df.sort_values("First Name",inplace=True)
emp_df.head(10)
```

```python
#calling len() before applying dropduplicates()
len(emp_df)
```

```python
#calling len() after applying dropduplicates()
emp_df.drop_duplicates()
len(emp_df)
```

```python
emp_df.drop_duplicates(subset=["First Name"]).head()
#by default it keeps the first occurence of the duplicate value and removes the remaining
```

```python
#keep=last
emp_df.drop_duplicates(subset=["First Name"],keep="last").head()
```

```python
#To completely remove the duplicates without keeping first or last occurrence
emp_df.drop_duplicates(subset=["First Name"],keep=False).head(10)
```

```python
#Mentioning multiple columns in subset parameter
emp_df.drop_duplicates(subset=["First Name","Team"],keep=False).head(10)
#Though we are seeing Adam two times they differ in Team
```
