# Datafrmae sorting

```python
import pandas as pd
nba=pd.read_csv("nba.csv")
#When we apply sort_values method to a dataframe, it will sort the entire dataframe by a given column
#Sorting a dataframe using salary column
nba.sort_values("Salary").head()
```

```python
#To sort it in descending order
nba.sort_values("Salary",ascending=False).head()
```

```python
#Usually null values present in the dataset will be at the end when we do sorting
nba.sort_values("Salary").tail()
```

```python
#We can make the null values to appear at the top
#inplace=True to make permanent changes to the dataframe
nba.sort_values("Salary",na_position="first",inplace=True)
nba.head()
#We can see that null values appear at the top now
```

# Sorting data frame using multiple columns

```python
nba.sort_values(["Team","Name"]).head(15)
#Above line will sort the dataframe first using the column Team then using Name
#Since we didn't mention anything for ascending by default both the columns will be sorted in ascending
```

```python
nba.sort_values(["Team","Name"],ascending=[True,False]).head(15)
#In this case first Team is sorted in ascending, then Name is sorted in descending
#We can give any number of columns and corresponding sorting order
```

# Sorting dataframe using sort_index method

```python
#To sort it in index descending order
nba.sort_index(ascending=False).head()
```

```python
#To sort it in index ascending order
nba.sort_index().head()
```
